```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/Linear Regression.csv')
```

```python
data.head()
```

```python
x = data['Hydrocarbon Level (X)']
y = data['Purity (Y)']
plt.figure()
sns.regplot(x=x, y=y, fit_reg=True)
plt.scatter(x.mean(), y.mean(), color="green")
plt.show()
```
